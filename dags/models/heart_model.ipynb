```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
df=pd.read_csv("data/Heart_Disease_Prediction.csv")
```

```python
X=df.iloc[:,1:14]
X
y=df['Heart Disease']
```

```python
# Import label encoder
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

y= label_encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from keras import Sequential #Helps to create Forward and backward propogation
from keras.layers import Dense #Helps to create neurons in ANN
from keras.layers import LeakyReLU,PReLU,ELU #activation functions
```

```python
classifier=Sequential()
classifier.add(Dense(units=11,activation='relu'))
classifier.add(Dense(units=7,activation='relu'))
classifier.add(Dense(units=6,activation='relu'))
classifier.add(Dense(units=1,activation='sigmoid'))
classifier.compile(optimizer='adam',loss="binary_crossentropy",metrics=["accuracy"])
```

```python
import tensorflow as tf
early_stopping=tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
)
```

```python
model_history=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=1000,callbacks=early_stopping)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Calculate the Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
```
